```python
# ============================================================
# SHL Grammar Scoring  WavLM + Whisper ASR + LightGBM Ensemble
# (Paths fixed + WavLM uses AutoFeatureExtractor instead of AutoProcessor)
# ============================================================

import os, gc, re, math, json, time, random, warnings
import numpy as np
import pandas as pd

from tqdm.auto import tqdm
tqdm.pandas()

import torch
import torchaudio

from transformers import AutoFeatureExtractor, AutoModel, pipeline
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

import lightgbm as lgb
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
torch.set_num_threads(2)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# -----------------------------
# Your directory structure
# -----------------------------
BASE_DIR = "/kaggle/input/shl-intern-hiring-assessment-2025/dataset"
TRAIN_CSV = os.path.join(BASE_DIR, "csvs/train.csv")
TEST_CSV  = os.path.join(BASE_DIR, "csvs/test.csv")
TRAIN_AUDIO_DIR = os.path.join(BASE_DIR, "audios/train")
TEST_AUDIO_DIR  = os.path.join(BASE_DIR, "audios/test")

assert os.path.exists(TRAIN_CSV), f"Missing {TRAIN_CSV}"
assert os.path.exists(TEST_CSV),  f"Missing {TEST_CSV}"
assert os.path.exists(TRAIN_AUDIO_DIR), f"Missing {TRAIN_AUDIO_DIR}"
assert os.path.exists(TEST_AUDIO_DIR),  f"Missing {TEST_AUDIO_DIR}"
def fix_name(x):
    return x if "." in x else x + ".wav"

df_train = pd.read_csv(TRAIN_CSV)
df_test  = pd.read_csv(TEST_CSV)
df_train["filename"] = df_train["filename"].apply(fix_name)
df_test["filename"]  = df_test["filename"].apply(fix_name)

print(df_train.head())
print(df_test.head())
# =============================
# 1) AUDIO LOADING
# =============================
TARGET_SR = 16000

def load_audio(path, target_sr=TARGET_SR):
    wav, sr = torchaudio.load(path)  # (C, T)
    if wav.size(0) > 1:
        wav = wav.mean(dim=0, keepdim=True)
    if sr != target_sr:
        wav = torchaudio.functional.resample(wav, sr, target_sr)
        sr = target_sr
    wav = wav.squeeze(0)  # (T,)
    peak = wav.abs().max().item()
    if peak > 0:
        wav = wav / peak
    return wav, sr

# =============================
# 2) WAVLM EMBEDDINGS (FROZEN)
# =============================
WAVLM_NAME = "microsoft/wavlm-base"

# IMPORTANT: Use AutoFeatureExtractor (not AutoProcessor)
feature_extractor = AutoFeatureExtractor.from_pretrained(WAVLM_NAME)
model_wavlm = AutoModel.from_pretrained(WAVLM_NAME).to(DEVICE)
model_wavlm.eval()
for p in model_wavlm.parameters():
    p.requires_grad = False

@torch.inference_mode()
def extract_wavlm_embedding(waveform_1d, sr=TARGET_SR, max_len_s=65):
    # Truncate long clips to keep memory stable
    max_len = int(max_len_s * sr)
    if waveform_1d.numel() > max_len:
        waveform_1d = waveform_1d[:max_len]

    inputs = feature_extractor(
        waveform_1d.numpy(),
        sampling_rate=sr,
        return_tensors="pt",
        padding=True
    )
    inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
    out = model_wavlm(**inputs)  # returns BaseModelOutput
    # Mean-pool time dimension
    h = out.last_hidden_state.mean(dim=1).squeeze(0)  # (hidden,)
    return h.detach().cpu().numpy()

def build_wavlm_embeddings(df, audio_dir, cache_path):
    if os.path.exists(cache_path):
        data = np.load(cache_path, allow_pickle=True)
        names = list(data["names"])
        embs = data["embs"]
        lookup = {names[i]: embs[i] for i in range(len(names))}
        return np.stack([lookup[f] for f in df["filename"]])

    embs, names = [], []
    for fn in tqdm(df["filename"], desc="WavLM embeddings"):
        path = os.path.join(audio_dir, fn)
        wav, sr = load_audio(path, TARGET_SR)
        vec = extract_wavlm_embedding(wav, sr)
        embs.append(vec); names.append(fn)

    X = np.stack(embs)
    np.savez_compressed(cache_path, names=np.array(names), embs=X)
    return X

print("Extracting WavLM embeddings (train)...")
X_audio_train = build_wavlm_embeddings(
    df_train, TRAIN_AUDIO_DIR, cache_path="/kaggle/working/wavlm_train.npz"
)
print("Extracting WavLM embeddings (test)...")
X_audio_test = build_wavlm_embeddings(
    df_test, TEST_AUDIO_DIR, cache_path="/kaggle/working/wavlm_test.npz"
)
y = df_train["label"].astype(float).values

# =============================
# 3) WHISPER ASR (TRANSCRIPTS)
# =============================
# If GPU is tight, you can switch to "openai/whisper-tiny" for speed.
ASR_MODEL = "openai/whisper-small"
asr_pipe = pipeline(
    "automatic-speech-recognition",
    model=ASR_MODEL,
    device=0 if DEVICE == "cuda" else -1,
    chunk_length_s=30,
    return_timestamps=False,
    generate_kwargs={"language": "en", "task": "transcribe"}
)

def transcribe_file(path):
    try:
        out = asr_pipe(path)
        if isinstance(out, dict) and "text" in out:
            return out["text"].strip()
        return ""
    except Exception:
        return ""

TRANSCRIPT_TRAIN_CACHE = "/kaggle/working/asr_train.json"
TRANSCRIPT_TEST_CACHE  = "/kaggle/working/asr_test.json"

def build_transcripts(df, audio_dir, cache):
    if os.path.exists(cache):
        return json.load(open(cache, "r"))
    res = {}
    for fn in tqdm(df["filename"], desc="ASR"):
        p = os.path.join(audio_dir, fn)
        res[fn] = transcribe_file(p)
    json.dump(res, open(cache, "w"))
    return res

print("Transcribing train...")
asr_train = build_transcripts(df_train, TRAIN_AUDIO_DIR, TRANSCRIPT_TRAIN_CACHE)
print("Transcribing test...")
asr_test  = build_transcripts(df_test,  TEST_AUDIO_DIR,  TRANSCRIPT_TEST_CACHE)

df_train["transcript"] = df_train["filename"].map(asr_train)
df_test["transcript"]  = df_test["filename"].map(asr_test)

# =============================
# 4) TEXT FEATURES (lightweight, robust)
# =============================
STOPWORDS = set("""
a an the is are am was were be been being i you he she it we they me my mine your yours his her its our their
this that these those to of for with in on at from by as and or but if then so because while though although
about into over after before under above between during without within not no nor do does did done doing have
has had having will would can could should may might must than too very more most less least also just only
""".strip().split())

DISFLUENCIES = ["um", "uh", "erm", "hmm", "like", "you know", "kind of", "sort of"]

def tokenize_words(s):
    return re.findall(r"[A-Za-z']+", s.lower())

def text_features_one(s):
    s = s.strip()
    toks = tokenize_words(s)
    ntok = len(toks)
    nuniq = len(set(toks))
    ttr = nuniq / ntok if ntok else 0.0

    sents = [x.strip() for x in re.split(r"[.!?]+", s) if x.strip()]
    nsent = len(sents)
    avg_sent_len = (ntok / nsent) if nsent else ntok

    nstop = sum(1 for t in toks if t in STOPWORDS)
    stop_ratio = nstop / ntok if ntok else 0.0

    s_low = s.lower()
    disfl_count = sum(s_low.count(d) for d in DISFLUENCIES)

    return dict(
        token_count=ntok,
        unique_ratio=ttr,
        sentence_count=nsent,
        avg_sentence_len=avg_sent_len,
        stopword_ratio=stop_ratio,
        disfluency_count=disfl_count,
    )

def build_text_feature_df(df, text_col="transcript"):
    rows = []
    for s in df[text_col].fillna(""):
        rows.append(text_features_one(s))
    return pd.DataFrame(rows).fillna(0.0)

X_text_train = build_text_feature_df(df_train)
X_text_test  = build_text_feature_df(df_test)

print("Audio emb shape:", X_audio_train.shape, "Text feat shape:", X_text_train.shape)

# =============================
# 5) LIGHTGBM MODELS + 5-FOLD CV
# =============================
def rmse(a, b):
    return math.sqrt(mean_squared_error(a, b))

def cv_lgbm_regression(X, y, n_splits=5, seed=SEED, params=None):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed)
    oof = np.zeros(len(y), dtype=float)
    models = []

    base = dict(
        objective="regression",
        metric="rmse",
        learning_rate=0.05,
        num_leaves=31,
        feature_fraction=0.9,
        bagging_fraction=0.8,
        bagging_freq=1,
        min_data_in_leaf=20,
        max_depth=-1,
        n_estimators=2000,
        random_state=seed,
        verbose=-1
    )
    if params: base.update(params)

    for fold, (trn_idx, val_idx) in enumerate(kf.split(X, y), 1):
        X_tr, X_va = X[trn_idx], X[val_idx]
        y_tr, y_va = y[trn_idx], y[val_idx]

        model = lgb.LGBMRegressor(**base)
        model.fit(
            X_tr, y_tr,
            eval_set=[(X_va, y_va)],
            eval_metric="rmse",
            callbacks=[lgb.early_stopping(stopping_rounds=100, verbose=False)]
        )
        pred = model.predict(X_va, num_iteration=model.best_iteration_)
        oof[val_idx] = pred
        print(f"[Fold {fold}] RMSE={rmse(y_va, pred):.4f} | Pearson={pearsonr(y_va, pred)[0]:.4f}")
        models.append(model)

    print(f"OOF RMSE={rmse(y, oof):.4f} | OOF Pearson={pearsonr(y, oof)[0]:.4f}")
    return models, oof

print("\nTraining LGBM on WavLM embeddings...")
mdl_audio, oof_audio = cv_lgbm_regression(X_audio_train, y, n_splits=5)

print("\nTraining LGBM on text features...")
mdl_text, oof_text = cv_lgbm_regression(X_text_train.values, y, n_splits=5)

# =============================
# 6) BLENDING (optimize on OOF)
# =============================
alphas = np.linspace(0, 1, 21)
best_alpha, best_rmse, best_p = None, 1e9, None
for a in alphas:
    oof_blend = a * oof_audio + (1 - a) * oof_text
    r = rmse(y, oof_blend)
    p = pearsonr(y, oof_blend)[0]
    if r < best_rmse:
        best_rmse, best_alpha, best_p = r, a, p

print(f"\nBest blend alpha (audio weight) = {best_alpha:.2f}")
print(f"Blended OOF RMSE = {best_rmse:.4f}")
print(f"Blended OOF Pearson = {best_p:.4f}")

print("\n=== REQUIRED METRIC ===")
print(f"Training OOF RMSE (to report): {best_rmse:.4f}")
print(f"Training OOF Pearson: {best_p:.4f}")

# Scatter plot (optional)
plt.figure(figsize=(5,5))
plt.scatter(y, oof_audio, s=12, alpha=0.6, label="Audio model")
plt.scatter(y, oof_text,  s=12, alpha=0.6, label="Text model")
plt.scatter(y, best_alpha*oof_audio + (1-best_alpha)*oof_text, s=16, alpha=0.8, label="Blend")
mn, mx = float(min(y.min(), oof_audio.min(), oof_text.min())), float(max(y.max(), oof_audio.max(), oof_text.max()))
plt.plot([mn, mx], [mn, mx], lw=1)
plt.xlabel("Ground truth"); plt.ylabel("Predicted"); plt.title("OOF Predictions vs Ground Truth")
plt.legend(); plt.show()

# =============================
# 7) TRAIN FINAL MODELS ON FULL DATA
# =============================
def fit_full_lgbm(X, y, n_estimators=800):
    model = lgb.LGBMRegressor(
        objective="regression",
        learning_rate=0.05,
        num_leaves=31,
        feature_fraction=0.95,
        bagging_fraction=0.9,
        bagging_freq=1,
        min_data_in_leaf=15,
        n_estimators=n_estimators,
        random_state=SEED,
        verbose=-1
    )
    model.fit(X, y)
    return model

print("\nFitting full LightGBM models...")
final_audio_model = fit_full_lgbm(X_audio_train, y)
final_text_model  = fit_full_lgbm(X_text_train.values, y)

# =============================
# 8) INFERENCE ON TEST & SUBMISSION
# =============================
def predict_with_ensemble(models, X):
    preds = np.zeros(X.shape[0], dtype=float)
    for m in models:
        preds += m.predict(X, num_iteration=getattr(m, "best_iteration_", None)) / len(models)
    return preds

test_pred_audio_cv = predict_with_ensemble(mdl_audio, X_audio_test)
test_pred_text_cv  = predict_with_ensemble(mdl_text,  X_text_test.values)

test_pred_audio_full = final_audio_model.predict(X_audio_test)
test_pred_text_full  = final_text_model.predict(X_text_test.values)

test_pred_audio = 0.5 * test_pred_audio_cv + 0.5 * test_pred_audio_full
test_pred_text  = 0.5 * test_pred_text_cv  + 0.5 * test_pred_text_full

test_blend = best_alpha * test_pred_audio + (1 - best_alpha) * test_pred_text
test_blend = np.clip(test_blend, 0.0, 5.0)

sub = pd.DataFrame({
    "filename": df_test["filename"],
    "label": test_blend
})
sub_path = "/kaggle/working/submission.csv"
sub.to_csv(sub_path, index=False)
print("Saved:", sub_path)
sub.head()
```

```python
# ============================================================
# SHL Grammar Scoring  Robust pipeline for BEST scores
# WavLM (+optional HuBERT) + MFCC/Prosody + Whisper text
# LLM signals: GPT-2 perplexity + DeBERTa-v3 PLL (auto-disable if unavailable)
# 5x LightGBM base models -> NNLS stack -> Isotonic calibration
# Prints REQUIRED Training OOF RMSE; saves submission.csv
# ============================================================

import os, re, gc, math, json, time, random, warnings
from typing import Dict, List
import numpy as np
import pandas as pd
from tqdm.auto import tqdm
tqdm.pandas()

import torch
import torchaudio
from torchaudio import transforms as T

import lightgbm as lgb
import matplotlib.pyplot as plt

from transformers import (
    AutoFeatureExtractor, AutoModel, AutoTokenizer,
    AutoModelForCausalLM, AutoModelForMaskedLM, pipeline
)
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.isotonic import IsotonicRegression
from scipy.stats import pearsonr, skew, kurtosis
from scipy.optimize import nnls

warnings.filterwarnings("ignore")

# -----------------------------
# Repro & Device
# -----------------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
torch.set_num_threads(2)
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# -----------------------------
# Paths (adjust if needed)
# -----------------------------
BASE_DIR = "/kaggle/input/shl-intern-hiring-assessment-2025/dataset"
TRAIN_CSV = os.path.join(BASE_DIR, "csvs/train.csv")
TEST_CSV  = os.path.join(BASE_DIR, "csvs/test.csv")
TRAIN_AUDIO_DIR = os.path.join(BASE_DIR, "audios/train")
TEST_AUDIO_DIR  = os.path.join(BASE_DIR, "audios/test")

assert os.path.exists(TRAIN_CSV), f"Missing {TRAIN_CSV}"
assert os.path.exists(TEST_CSV),  f"Missing {TEST_CSV}"
assert os.path.exists(TRAIN_AUDIO_DIR), f"Missing {TRAIN_AUDIO_DIR}"
assert os.path.exists(TEST_AUDIO_DIR),  f"Missing {TEST_AUDIO_DIR}"

def fix_name(x):  # ensure .wav suffix if missing
    return x if "." in x else x + ".wav"

df_train = pd.read_csv(TRAIN_CSV)
df_test  = pd.read_csv(TEST_CSV)
df_train["filename"] = df_train["filename"].apply(fix_name)
df_test["filename"]  = df_test["filename"].apply(fix_name)
y = df_train["label"].astype(float).values

print("Train:", df_train.shape, " Test:", df_test.shape)
print(df_train.head(3))

# ============================================================
# 1) AUDIO LOADING (ROBUST) & LIGHT CLEANING
# ============================================================
TARGET_SR = 16000
MIN_SEC = 0.25
MIN_SAMPLES = int(MIN_SEC * TARGET_SR)

def _simple_trim(wav: torch.Tensor, thr: float = 0.005, pad: int = 1600):
    # energy-based trim with small padding
    if wav.numel() == 0: return wav
    x = wav.abs().cpu().numpy()
    idx = np.where(x > thr)[0]
    if len(idx) == 0:
        return wav
    lo, hi = max(idx[0]-pad, 0), min(idx[-1]+pad, wav.numel()-1)
    return wav[lo:hi+1]

def _ensure_min_len(wav: torch.Tensor, min_samples: int = MIN_SAMPLES):
    if wav.numel() >= min_samples:
        return wav
    pad = min_samples - wav.numel()
    if pad <= 0: 
        return wav
    return torch.nn.functional.pad(wav, (0, pad), mode="constant", value=0.0)

def load_audio(path, target_sr=TARGET_SR):
    # Load mono
    try:
        wav, sr = torchaudio.load(path)  # (C, T)
    except Exception as e:
        # Corrupt/missing: return short silence
        wav = torch.zeros(target_sr, dtype=torch.float32); sr = target_sr

    if wav.size(0) > 1:
        wav = wav.mean(dim=0, keepdim=True)
    wav = wav.squeeze(0)  # (T,)

    # Resample
    if sr != target_sr and wav.numel() > 0:
        wav = torchaudio.functional.resample(wav, sr, target_sr)

    orig = wav.clone()

    # Try VAD
    try:
        if wav.numel() > 0:
            wav = torchaudio.functional.vad(wav, target_sr)
    except Exception:
        pass

    # If VAD made it empty or too short, try simple trim
    if wav.numel() < MIN_SAMPLES:
        wav = _simple_trim(orig, thr=0.005)

    # If still too short/empty, fallback to original
    if wav.numel() < MIN_SAMPLES:
        wav = orig

    # If still too short, pad with zeros
    wav = _ensure_min_len(wav, MIN_SAMPLES)

    # Peak normalize if non-zero
    if wav.numel() > 0:
        peak = float(wav.abs().max().item()) if wav.numel() else 0.0
        if peak > 0:
            wav = wav / peak

    return wav, target_sr

# ============================================================
# 2) SSL: WavLM & optional HuBERT Embeddings (FROZEN, RESILIENT)
# ============================================================
def safe_load_ssl_model(model_name):
    try:
        feat = AutoFeatureExtractor.from_pretrained(model_name)
        mdl  = AutoModel.from_pretrained(model_name).to(DEVICE).eval()
        for p in mdl.parameters(): p.requires_grad = False
        return feat, mdl
    except Exception as e:
        print(f"[WARN] Could not load {model_name}: {e}. Disabling this branch.")
        return None, None

def extract_ssl_embeddings(model_name: str, df: pd.DataFrame, audio_dir: str, cache_path: str):
    if os.path.exists(cache_path):
        data = np.load(cache_path, allow_pickle=True)
        names = list(data["names"]); embs = data["embs"]
        lookup = {names[i]: embs[i] for i in range(len(names))}
        return np.stack([lookup[f] for f in df["filename"]]).astype(np.float32)

    feat, mdl = safe_load_ssl_model(model_name)
    if feat is None or mdl is None:
        # Return zeros to keep shapes consistent in concat
        print(f"[INFO] Using zero embeddings for {model_name}")
        # Discover dim via a tiny fake forward if possible; otherwise 768 fallback
        dim = 768
        try:
            dummy = torch.zeros(TARGET_SR, dtype=torch.float32)
            inputs = feat(dummy.numpy(), sampling_rate=TARGET_SR, return_tensors="pt", padding=True)
            inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
            with torch.inference_mode():
                out = mdl(**inputs)
                dim = int(out.last_hidden_state.shape[-1])
        except Exception:
            pass
        return np.zeros((len(df), dim), dtype=np.float32)

    # helper to featurize one sample (with safety net)
    def featurize_one(waveform_1d: torch.Tensor, sr=TARGET_SR, max_len_s=65):
        max_len = int(max_len_s * sr)
        if waveform_1d.numel() > max_len:
            waveform_1d = waveform_1d[:max_len]
        inputs = feat(
            waveform_1d.cpu().numpy().astype(np.float32),
            sampling_rate=sr,
            return_tensors="pt",
            padding=True
        )
        inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
        with torch.inference_mode():
            out = mdl(**inputs)  # BaseModelOutput
            h = out.last_hidden_state.mean(dim=1).squeeze(0)
        return h.detach().cpu().numpy().astype(np.float32)

    # Discover embedding dim once (for graceful fallback)
    try:
        _dim_probe = featurize_one(torch.zeros(TARGET_SR))
        embed_dim = int(_dim_probe.shape[0])
    except Exception:
        embed_dim = 768

    embs, names = [], []
    failures = 0
    for fn in tqdm(df["filename"], desc=f"{model_name} embeddings"):
        p = os.path.join(audio_dir, fn)
        try:
            wav, sr = load_audio(p, TARGET_SR)
            vec = featurize_one(wav, sr)
        except Exception as e:
            failures += 1
            vec = np.zeros(embed_dim, dtype=np.float32)  # safe fallback
        embs.append(vec); names.append(fn)

    if failures:
        print(f"[WARN] {model_name}: {failures} files fell back to zeros.")

    X = np.stack(embs).astype(np.float32)
    np.savez_compressed(cache_path, names=np.array(names), embs=X)
    return X

print("Extracting SSL embeddings ")
X_wavlm_train = extract_ssl_embeddings("microsoft/wavlm-base", df_train, TRAIN_AUDIO_DIR, "/kaggle/working/wavlm_train.npz")
X_wavlm_test  = extract_ssl_embeddings("microsoft/wavlm-base", df_test,  TEST_AUDIO_DIR,  "/kaggle/working/wavlm_test.npz")
X_hubert_train = extract_ssl_embeddings("facebook/hubert-base-ls960", df_train, TRAIN_AUDIO_DIR, "/kaggle/working/hubert_train.npz")
X_hubert_test  = extract_ssl_embeddings("facebook/hubert-base-ls960", df_test,  TEST_AUDIO_DIR,  "/kaggle/working/hubert_test.npz")
print("SSL shapes:", X_wavlm_train.shape, X_hubert_train.shape)

# ============================================================
# 3) MFCC / PROSODY FEATURES (ROBUST)
# ============================================================
_mfcc = T.MFCC(
    sample_rate=TARGET_SR, n_mfcc=40,
    melkwargs=dict(n_fft=1024, hop_length=256, n_mels=64, f_min=20, f_max=8000)
)
_delta = T.ComputeDeltas(win_length=5)

def zcr(x: torch.Tensor):
    x = x.view(-1)
    if x.numel() < 2:
        return 0.0
    return float(((x[1:] * x[:-1]) < 0).float().mean().item())

def summarize_stats(x_np: np.ndarray, prefix: str):
    feats = {}
    if x_np.size == 0:
        feats[f"{prefix}_mean"] = 0.0
        feats[f"{prefix}_std"] = 0.0
        feats[f"{prefix}_skew"] = 0.0
        feats[f"{prefix}_kurt"] = 0.0
        return feats
    if x_np.ndim == 1:
        arr = x_np
        feats[f"{prefix}_mean"] = float(np.mean(arr))
        feats[f"{prefix}_std"]  = float(np.std(arr))
        feats[f"{prefix}_skew"] = float(skew(arr, bias=False)) if arr.size>3 else 0.0
        feats[f"{prefix}_kurt"] = float(kurtosis(arr, bias=False)) if arr.size>3 else 0.0
        return feats
    mean_ = x_np.mean(axis=1); std_ = x_np.std(axis=1)
    skew_ = np.array([skew(x_np[i], bias=False) if x_np.shape[1]>3 else 0.0 for i in range(x_np.shape[0])])
    kurt_ = np.array([kurtosis(x_np[i], bias=False) if x_np.shape[1]>3 else 0.0 for i in range(x_np.shape[0])])
    for arr, tag in [(mean_, "mean"), (std_, "std"), (skew_, "skew"), (kurt_, "kurt")]:
        feats.update({f"{prefix}_{tag}_{j}": float(v) for j, v in enumerate(arr)})
    return feats

def mfcc_feature_row(wav: torch.Tensor, sr=TARGET_SR) -> Dict[str, float]:
    wav = _ensure_min_len(wav, MIN_SAMPLES)
    dur_s = wav.numel() / sr
    rms = float(torch.sqrt(torch.clamp(wav.pow(2).mean(), 1e-12, None)))
    zero_cr = zcr(wav)
    thr = 0.01
    sil_ratio = float((wav.abs() < thr).float().mean().item())
    with torch.no_grad():
        try:
            m = _mfcc(wav.unsqueeze(0)).squeeze(0)  # (n_mfcc, T)
            d = _delta(m); dd = _delta(d)
        except Exception:
            m = torch.zeros(40, 1); d = torch.zeros(40, 1); dd = torch.zeros(40, 1)
    feats = {}
    feats.update(summarize_stats(m.cpu().numpy(), "mfcc"))
    feats.update(summarize_stats(d.cpu().numpy(), "delta"))
    feats.update(summarize_stats(dd.cpu().numpy(), "delta2"))
    feats.update(dict(duration_s=dur_s, rms=rms, zcr=zero_cr, silence_ratio=sil_ratio))
    return feats

def build_mfcc_df(df, audio_dir, cache_json):
    if os.path.exists(cache_json):
        return pd.read_json(cache_json, orient="records")
    rows = []
    for fn in tqdm(df["filename"], desc="MFCC/Prosody"):
        p = os.path.join(audio_dir, fn)
        wav, sr = load_audio(p, TARGET_SR)
        row = mfcc_feature_row(wav, sr)
        row["filename"] = fn
        rows.append(row)
    out = pd.DataFrame(rows)
    out.to_json(cache_json, orient="records")
    return out

mfcc_train = build_mfcc_df(df_train, TRAIN_AUDIO_DIR, "/kaggle/working/mfcc_train.json")
mfcc_test  = build_mfcc_df(df_test,  TEST_AUDIO_DIR,  "/kaggle/working/mfcc_test.json")
mfcc_train = mfcc_train.set_index("filename").reindex(df_train["filename"]).reset_index()
mfcc_test  = mfcc_test.set_index("filename").reindex(df_test["filename"]).reset_index()
feat_cols_mfcc = [c for c in mfcc_train.columns if c != "filename"]
X_mfcc_train = mfcc_train[feat_cols_mfcc].astype(np.float32).values
X_mfcc_test  = mfcc_test[feat_cols_mfcc].astype(np.float32).values
print("MFCC/Prosody:", X_mfcc_train.shape, X_mfcc_test.shape)

# ============================================================
# 4) ASR TRANSCRIPTS + LLM TEXT FEATURES (auto-disable if unavailable)
# ============================================================
ASR_MODEL = "openai/whisper-small"  # change to "openai/whisper-tiny" if runtime tight
TRANSCRIPT_TRAIN_CACHE = "/kaggle/working/asr_train.json"
TRANSCRIPT_TEST_CACHE  = "/kaggle/working/asr_test.json"

if os.path.exists(TRANSCRIPT_TRAIN_CACHE) and os.path.exists(TRANSCRIPT_TEST_CACHE):
    asr_train = json.load(open(TRANSCRIPT_TRAIN_CACHE, "r"))
    asr_test  = json.load(open(TRANSCRIPT_TEST_CACHE, "r"))
else:
    try:
        asr_pipe = pipeline(
            "automatic-speech-recognition",
            model=ASR_MODEL,
            device=0 if DEVICE == "cuda" else -1,
            chunk_length_s=30,
            return_timestamps=False,
            generate_kwargs={"language": "en", "task": "transcribe"}
        )
        def transcribe_file(path):
            try:
                out = asr_pipe(path)
                if isinstance(out, dict) and "text" in out:
                    return out["text"].strip()
                return ""
            except Exception:
                return ""
        def build_transcripts(df, audio_dir):
            res = {}
            for fn in tqdm(df["filename"], desc="ASR"):
                p = os.path.join(audio_dir, fn)
                res[fn] = transcribe_file(p)
            return res
        asr_train = build_transcripts(df_train, TRAIN_AUDIO_DIR)
        asr_test  = build_transcripts(df_test,  TEST_AUDIO_DIR)
    except Exception as e:
        print(f"[WARN] Whisper ASR unavailable: {e}. Using empty transcripts.")
        asr_train = {fn: "" for fn in df_train["filename"]}
        asr_test  = {fn: "" for fn in df_test["filename"]}
    json.dump(asr_train, open(TRANSCRIPT_TRAIN_CACHE, "w"))
    json.dump(asr_test,  open(TRANSCRIPT_TEST_CACHE,  "w"))

df_train["transcript"] = df_train["filename"].map(asr_train).fillna("")
df_test["transcript"]  = df_test["filename"].map(asr_test).fillna("")

STOPWORDS = set("""
a an the is are am was were be been being i you he she it we they me my mine your yours his her its our their
this that these those to of for with in on at from by as and or but if then so because while though although
about into over after before under above between during without within not no nor do does did done doing have
has had having will would can could should may might must than too very more most less least also just only
""".strip().split())
DISFLUENCIES = ["um", "uh", "erm", "hmm", "like", "you know", "kind of", "sort of"]
def tokenize_words(s): return re.findall(r"[A-Za-z']+", s.lower())

# --- GPT-2 perplexity (avg NLL) ---
PPL_CACHE_TRAIN = "/kaggle/working/ppl_train.json"
PPL_CACHE_TEST  = "/kaggle/working/ppl_test.json"
ppl_train, ppl_test = {}, {}
try:
    if os.path.exists(PPL_CACHE_TRAIN) and os.path.exists(PPL_CACHE_TEST):
        ppl_train = json.load(open(PPL_CACHE_TRAIN, "r"))
        ppl_test  = json.load(open(PPL_CACHE_TEST, "r"))
    else:
        tok_gpt2 = AutoTokenizer.from_pretrained("gpt2")
        model_gpt2 = AutoModelForCausalLM.from_pretrained("gpt2")
        model_gpt2.eval()  # CPU is fine
        def text_avg_nll(text: str, max_len=512, stride=384):
            if not text or not text.strip():
                return 5.0
            ids = tok_gpt2(text, return_tensors="pt")["input_ids"][0]
            nlls = []
            for i in range(0, len(ids), stride):
                chunk = ids[i:i+max_len]
                if chunk.numel() < 2: break
                inp = chunk.unsqueeze(0)
                with torch.no_grad():
                    out = model_gpt2(inp, labels=inp)
                nlls.append(float(out.loss.item()))
                if i + max_len >= len(ids): break
            return float(np.mean(nlls)) if nlls else 5.0
        ppl_train = {fn: text_avg_nll(asr_train.get(fn, "")) for fn in tqdm(df_train["filename"], desc="GPT-2 PPL train")}
        ppl_test  = {fn: text_avg_nll(asr_test.get(fn, ""))   for fn in tqdm(df_test["filename"],  desc="GPT-2 PPL test")}
        json.dump(ppl_train, open(PPL_CACHE_TRAIN, "w"))
        json.dump(ppl_test,  open(PPL_CACHE_TEST,  "w"))
except Exception as e:
    print(f"[WARN] GPT-2 PPL disabled: {e}")

# --- DeBERTa-v3 masked-LM pseudo log-likelihood (grammar proxy) ---
PLL_CACHE_TRAIN = "/kaggle/working/pll_train.json"
PLL_CACHE_TEST  = "/kaggle/working/pll_test.json"
pll_train, pll_test = {}, {}
try:
    if os.path.exists(PLL_CACHE_TRAIN) and os.path.exists(PLL_CACHE_TEST):
        pll_train = json.load(open(PLL_CACHE_TRAIN, "r"))
        pll_test  = json.load(open(PLL_CACHE_TEST, "r"))
    else:
        tok_mask = AutoTokenizer.from_pretrained("microsoft/deberta-v3-base")
        model_mask = AutoModelForMaskedLM.from_pretrained("microsoft/deberta-v3-base")
        model_mask.eval()
        def masked_token_logprob(input_ids, pos):
            masked = input_ids.clone()
            masked[pos] = tok_mask.mask_token_id
            with torch.no_grad():
                out = model_mask(masked.unsqueeze(0))
                logits = out.logits[0, pos]
                lp = torch.log_softmax(logits, dim=-1)
            return float(lp[input_ids[pos]].item())
        def pseudo_log_likelihood(text: str, max_tokens=256, step=2):
            if not text or not text.strip():
                return -3.0
            enc = tok_mask(text, return_tensors="pt", truncation=True, max_length=max_tokens)
            ids = enc["input_ids"][0]
            positions = list(range(1, ids.size(0)-1, step))
            if len(positions) == 0:
                return -3.0
            lps = [masked_token_logprob(ids, p) for p in positions]
            return float(np.mean(lps))
        pll_train = {fn: pseudo_log_likelihood(asr_train.get(fn, ""), max_tokens=256, step=2)
                     for fn in tqdm(df_train["filename"], desc="DeBERTa PLL train")}
        pll_test  = {fn: pseudo_log_likelihood(asr_test.get(fn, ""),  max_tokens=256, step=2)
                     for fn in tqdm(df_test["filename"],  desc="DeBERTa PLL test")}
        json.dump(pll_train, open(PLL_CACHE_TRAIN, "w"))
        json.dump(pll_test,  open(PLL_CACHE_TEST,  "w"))
except Exception as e:
    print(f"[WARN] DeBERTa PLL disabled: {e}")

def text_features_one(text: str, dur_s: float, fn: str):
    s = (text or "").strip()
    toks = tokenize_words(s)
    ntok = len(toks); nuniq = len(set(toks))
    ttr = nuniq / ntok if ntok else 0.0
    sents = [x.strip() for x in re.split(r"[.!?]+", s) if x.strip()]
    nsent = len(sents)
    avg_sent_len = (ntok / nsent) if nsent else ntok
    nstop = sum(1 for t in toks if t in STOPWORDS)
    stop_ratio = nstop / ntok if ntok else 0.0
    s_low = s.lower()
    disfl_count = sum(s_low.count(d) for d in DISFLUENCIES)
    bigrams = [" ".join(toks[i:i+2]) for i in range(max(ntok-1, 0))]
    rep_rate = 0.0
    if bigrams:
        from collections import Counter
        cnt = Counter(bigrams)
        rep_rate = sum(v-1 for v in cnt.values() if v>1) / len(bigrams)
    speaking_rate = (ntok / max(dur_s, 1e-6)) if dur_s>0 else 0.0
    ppl = float(ppl_train.get(fn, 5.0)) if ppl_train else 5.0
    pll = float(pll_train.get(fn, -3.0)) if pll_train else -3.0
    return dict(
        token_count=ntok,
        unique_ratio=ttr,
        sentence_count=nsent,
        avg_sentence_len=avg_sent_len,
        stopword_ratio=stop_ratio,
        disfluency_count=disfl_count,
        repetition_rate=rep_rate,
        speaking_rate=speaking_rate,
        char_len=len(s),
        gpt2_avg_nll=ppl,          # lower is better
        deberta_pll=pll            # higher (less negative) is better
    )

def build_text_feature_df(df, mfcc_df):
    dur_lookup = dict(zip(mfcc_df["filename"], mfcc_df["duration_s"]))
    rows = []
    for fn, s in zip(df["filename"], df["transcript"]):
        rows.append(text_features_one(s, float(dur_lookup.get(fn, 0.0)), fn))
    return pd.DataFrame(rows).fillna(0.0)

X_text_train_df = build_text_feature_df(df_train, mfcc_train)
X_text_test_df  = build_text_feature_df(df_test,  mfcc_test)
# replace cached LLM vals for test (if available)
if ppl_test:
    X_text_test_df["gpt2_avg_nll"] = [float(ppl_test.get(fn, 5.0)) for fn in df_test["filename"]]
if pll_test:
    X_text_test_df["deberta_pll"]  = [float(pll_test.get(fn, -3.0)) for fn in df_test["filename"]]

X_text_train = X_text_train_df.values.astype(np.float32)
X_text_test  = X_text_test_df.values.astype(np.float32)
print("Text feats:", X_text_train.shape, X_text_test.shape)

# ============================================================
# 5) LightGBM CV per branch
# ============================================================
def rmse(a, b): return math.sqrt(mean_squared_error(a, b))

def cv_lgbm(X, y, name, params=None, n_splits=5, seed=SEED):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed)
    oof = np.zeros(len(y), dtype=np.float32)
    models = []
    base = dict(
        objective="regression",
        metric="rmse",
        learning_rate=0.03,
        num_leaves=63,
        feature_fraction=0.9,
        bagging_fraction=0.8,
        bagging_freq=1,
        min_data_in_leaf=15,
        max_depth=-1,
        n_estimators=5000,
        random_state=seed,
        n_jobs=-1,
        verbose=-1
    )
    if params: base.update(params)

    for fold, (tr, va) in enumerate(kf.split(X, y), 1):
        X_tr, y_tr = X[tr], y[tr]
        X_va, y_va = X[va], y[va]
        model = lgb.LGBMRegressor(**base)
        model.fit(
            X_tr, y_tr,
            eval_set=[(X_va, y_va)],
            eval_metric="rmse",
            callbacks=[lgb.early_stopping(stopping_rounds=300, verbose=False)]
        )
        pred = model.predict(X_va, num_iteration=model.best_iteration_)
        oof[va] = pred.astype(np.float32)
        print(f"[{name} | Fold {fold}] RMSE={rmse(y_va, pred):.4f} | Pearson={pearsonr(y_va, pred)[0]:.4f}")
        models.append(model)
    print(f"[{name}] OOF RMSE={rmse(y, oof):.4f} | OOF Pearson={pearsonr(y, oof)[0]:.4f}")
    return models, oof

print("\nTraining base models ")
mdl_wavlm,  oof_wavlm  = cv_lgbm(X_wavlm_train, y, "WavLM")
mdl_hubert, oof_hubert = cv_lgbm(X_hubert_train, y, "HuBERT")
mdl_mfcc,   oof_mfcc   = cv_lgbm(X_mfcc_train, y, "MFCC")
mdl_text,   oof_text   = cv_lgbm(X_text_train, y, "TEXT")

X_concat_train = np.concatenate([X_wavlm_train, X_hubert_train, X_mfcc_train, X_text_train], axis=1)
X_concat_test  = np.concatenate([X_wavlm_test,  X_hubert_test,  X_mfcc_test,  X_text_test],  axis=1)
mdl_concat, oof_concat = cv_lgbm(
    X_concat_train, y, "CONCAT",
    params=dict(num_leaves=127, feature_fraction=0.75, learning_rate=0.02, n_estimators=8000)
)

# ============================================================
# 6) OOF Stacking (NNLS) + Isotonic Calibration
# ============================================================
oof_stack = np.vstack([oof_wavlm, oof_hubert, oof_mfcc, oof_text, oof_concat]).T  # (N, 5)
w, _ = nnls(oof_stack, y)  # non-negative weights
if w.sum() > 0: w = w / w.sum()
else: w = np.array([1.0, 0, 0, 0, 0], dtype=np.float32)
oof_blend = (oof_stack @ w)
oof_rmse  = rmse(y, oof_blend)
oof_p     = pearsonr(y, oof_blend)[0]
print(f"\n[STACK] Weights = WavLM:{w[0]:.3f} HuBERT:{w[1]:.3f} MFCC:{w[2]:.3f} TEXT:{w[3]:.3f} CONCAT:{w[4]:.3f}")
print(f"[STACK] OOF RMSE={oof_rmse:.4f} | OOF Pearson={oof_p:.4f}")

iso = IsotonicRegression(y_min=0.0, y_max=5.0, increasing=True, out_of_bounds="clip")
iso.fit(oof_blend, y)
oof_blend_iso = iso.transform(oof_blend)
oof_rmse_iso  = rmse(y, oof_blend_iso)
oof_p_iso     = pearsonr(y, oof_blend_iso)[0]
print(f"[STACK+ISO] OOF RMSE={oof_rmse_iso:.4f} | OOF Pearson={oof_p_iso:.4f}")

print("\n=== REQUIRED METRIC ===")
print(f"Training OOF RMSE (pre-iso):  {oof_rmse:.4f}")
print(f"Training OOF RMSE (post-iso): {oof_rmse_iso:.4f}")

# (Optional) OOF plot
try:
    plt.figure(figsize=(5,5))
    lo, hi = float(min(y.min(), oof_blend.min())), float(max(y.max(), oof_blend.max()))
    plt.scatter(y, oof_wavlm,  s=10, alpha=0.35, label="WavLM")
    plt.scatter(y, oof_hubert, s=10, alpha=0.35, label="HuBERT")
    plt.scatter(y, oof_mfcc,   s=10, alpha=0.35, label="MFCC")
    plt.scatter(y, oof_text,   s=10, alpha=0.35, label="TEXT")
    plt.scatter(y, oof_concat, s=10, alpha=0.35, label="CONCAT")
    plt.scatter(y, oof_blend,  s=18, alpha=0.8,  label="STACK")
    plt.plot([lo,hi],[lo,hi], lw=1)
    plt.xlabel("Ground truth"); plt.ylabel("Predicted"); plt.title("OOF vs Ground Truth")
    plt.legend(); plt.show()
except Exception as e:
    print("Plot skipped:", e)

# ============================================================
# 7) Train full-data models & predict test
# ============================================================
def fit_full(X, y, params):
    m = lgb.LGBMRegressor(**params)
    m.fit(X, y)
    return m

base_params = dict(
    objective="regression",
    learning_rate=0.03,
    num_leaves=63,
    feature_fraction=0.95,
    bagging_fraction=0.9,
    bagging_freq=1,
    min_data_in_leaf=10,
    n_estimators=5000,
    random_state=SEED,
    n_jobs=-1,
    verbose=-1
)
concat_params = dict(
    objective="regression",
    learning_rate=0.02,
    num_leaves=127,
    feature_fraction=0.75,
    bagging_fraction=0.9,
    bagging_freq=1,
    min_data_in_leaf=10,
    n_estimators=8000,
    random_state=SEED,
    n_jobs=-1,
    verbose=-1
)

print("\nFitting full-data models ")
final_wavlm  = fit_full(X_wavlm_train,  y, base_params)
final_hubert = fit_full(X_hubert_train, y, base_params)
final_mfcc   = fit_full(X_mfcc_train,   y, base_params)
final_text   = fit_full(X_text_train,   y, base_params)
final_concat = fit_full(X_concat_train, y, concat_params)

def predict_ensemble(models, Xs):
    return np.column_stack([m.predict(X) for m, X in zip(models, Xs)])

test_cols = predict_ensemble(
    [final_wavlm, final_hubert, final_mfcc, final_text, final_concat],
    [X_wavlm_test, X_hubert_test, X_mfcc_test, X_text_test, X_concat_test]
)  # (T, 5)

test_blend = test_cols @ w
test_blend = iso.transform(test_blend)
test_blend = np.clip(test_blend, 0.0, 5.0)

sub = pd.DataFrame({"filename": df_test["filename"], "label": test_blend})
sub_path = "/kaggle/working/submission.csv"
sub.to_csv(sub_path, index=False)
print("Saved:", sub_path)
try:
    display(sub.head())
except Exception:
    print(sub.head())

# Cleanup
gc.collect()
```
